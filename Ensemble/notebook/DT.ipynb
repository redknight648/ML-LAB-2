```python
from sklearn import tree
from sklearn.model_selection import train_test_split
import pandas as pd

data = pd.read_csv('rain_tree.csv')
data
X = data[['Humidity9am','RainToday']]
Y = data[['RainTomorrow']]

model = tree.DecisionTreeClassifier()
model = model.fit(X, Y)
tree.plot_tree(model)
```
