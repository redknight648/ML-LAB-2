```python
import pandas as pd

data = {'YouTube': [100, 50, 125, 75, 150],
        'Google': [50, 125, 100, 100, 50],
       'Residual': [-80.9, -37.8, 99.4, -12.3, 30.8]}
  

df = pd.DataFrame(data, columns=['YouTube', 'Google','Residual'])
df
```

```python
X = df[['YouTube','Google']]
Y = df[['Residual']]

from sklearn.linear_model import LinearRegression

regr = LinearRegression()  
regr.fit(X, Y)
df['y_pred'] = regr.predict(X)
df
```
