```python
yes_no_cols = ['Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup',
               'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
```

```python
data.head()
```

```python
for col in yes_no_cols:
    data[col].replace({'Yes':1,'No':0},inplace = True)
data.head()
```

```python
for col in data:
    print(f'{col}:{data[col].unique()}')
```

```python
data['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
dummyfied_data = pd.get_dummies(data = data, columns = ['Contract','PaymentMethod'])
dummyfied_data.head()
```

```python
dummyfied_data.dtypes
```

```python
col_to_scale =['tenure','MonthlyCharges','MonthlyCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

dummyfied_data[col_to_scale] = scaler.fit_transform(dummyfied_data[col_to_scale])
```

```python
dummyfied_data.sample(3)
```

```python
X = dummyfied_data.drop('Churn',axis = 1)
Y = dummyfied_data['Churn']
```

```python
from sklearn.model_selection import train_test_split

# train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.20, random_state = 42)
test_df = pd.concat ([X_test,y_test],axis =1 )
```

```python
X_train.shape
```

```python
len(X_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
        keras.layers.Dense(20, input_shape = (24,),activation = 'relu'),
        keras.layers.Dense(10, input_shape = (20,),activation = 'relu'),
        keras.layers.Dense(5, input_shape = (10,),activation = 'relu'),
        keras.layers.Dense(1,activation = 'sigmoid'),
])

model.compile(optimizer ='adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy']
             )

model.fit(X_train,y_train,epochs = 5)
```

```python
model.evaluate(X_test,y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
        
y_pred[:20]
```

```python
y_test[:20]
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sn

cm = confusion_matrix(y_test, y_pred)

cm_matrix = pd.DataFrame(data=cm, columns=['Actual Churn:1', 'Actual Churn:0'], 
                                 index=['Predict Churn:1', 'Predict Churn:0'])

sn.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```
